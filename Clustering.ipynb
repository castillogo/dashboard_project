## Preprocessing

**Curse of Dimensionality:** Clustering does not work very well with 10+ features if you use an Euclidean Distance

--> reduce number of features

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
pca = PCA(n_components=20)
Xt = pca.fit_transform(X)
Xt.shape
```

```python
plt.bar(range(20), np.cumsum(pca.explained_variance_ratio_))
```

```python
# plot 2 principal components
plt.scatter(Xt[:,0], Xt[:,1], c=y)
```
