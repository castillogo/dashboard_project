## Preprocessing

**Curse of Dimensionality:** Clustering does not work very well with 10+ features if you use an Euclidean Distance

--> reduce number of features

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
pca = PCA(n_components=20)
Xt = pca.fit_transform(X)
Xt.shape
```

```python
plt.bar(range(20), np.cumsum(pca.explained_variance_ratio_))
```

```python
# plot 2 principal components
plt.scatter(Xt[:,0], Xt[:,1], c=y)
```

When you do clustering: features must be comparable -> scale the data

```python
from sklearn.preprocessing import MinMaxScaler

Xtsc = MinMaxScaler().fit_transform(Xt)
```

### K-Means Clustering
* you have to decide # of clusters before
* assumption: clusters are spherical

```python
from sklearn.cluster import KMeans
```

```python
m = KMeans(n_clusters=10, random_state=42, tol=0.01)  # naive guess 10 because we have 10 digits
m.fit(Xtsc)
```

```python
clusters = m.predict(Xtsc)
clusters[:20]
```

```python
clusters = m.predict(Xtsc)
clusters[:20]  # output are cluster labels
```

```python
plt.scatter(Xt[:,0], Xt[:,1], c=clusters)
```

```python
cluster5 = images[clusters == 5]
```

```python
cluster5.shape
```

```python
plt.imshow(cluster5[8])
```

```python
labeldict = dict(zip(range(10), 'ABCDEFGHIJ'))
```

```python
clusters.shape, y.shape
```

```python
df = pd.DataFrame({'y': y, 'cluster':clusters})
df['cluster'].replace(labeldict, inplace=True)
count = df.groupby('y')['cluster'].value_counts().unstack().fillna(0.0)
sns.heatmap(count)
```

### t-SNE
* belongs to "manifold learning"
* tries to preserve distances between points
* does a dimensionality reduction from <50 features to 2
* good for visualizing data
* slower
* non-linear transformation

```python
from sklearn.manifold import TSNE
```

```python
Xsne = TSNE().fit_transform(Xtsc)
Xsne.shape
```

```python
plt.scatter(Xsne[:,0], Xsne[:,1], c=y)
```

```python
m = KMeans(n_clusters=10, random_state=42, tol=0.01)  # naive guess 10 because we have 10 digits
m.fit(Xsne)
clusters = m.predict(Xsne)
plt.scatter(Xsne[:,0], Xsne[:,1], c=clusters)
plt.plot(m.cluster_centers_[:,0], m.cluster_centers_[:,1], 'ro')
```

```python
df = pd.DataFrame({'y': y, 'cluster':clusters})
df['cluster'].replace(labeldict, inplace=True)
count = df.groupby('y')['cluster'].value_counts().unstack().fillna(0.0)
sns.heatmap(count)
```
